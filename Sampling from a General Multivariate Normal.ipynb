```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cov = np.array([[1, 0.8],[0.8, 3]])
```

```python
from scipy.stats import multivariate_normal as mvn
```

```python
mu = np.array([0,2])
```

```python
r = mvn.rvs(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0], r[:,1])
```

```python
plt.axis('equal')
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0],r[:,1])
```
